```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load public free model
tokenizer = AutoTokenizer.from_pretrained("Salesforce/codegen-350M-mono")
model = AutoModelForCausalLM.from_pretrained("Salesforce/codegen-350M-mono", device_map="auto")
model.eval()

# Fix padding issue
tokenizer.pad_token = tokenizer.eos_token


```

```python
model.eval()
```

```python
def detect_bug(code):
    if "/" in code and "0" in code:
        return "Bug Detected"
    if "def" in code and "return" not in code:
        return "Potential Logical Bug"
    return "No Bug Found"

```

```python
def generate_explanation(code, issue):
    prompt = f"""
You are a senior software engineer.
Issue Type: {issue}

Analyze the following code and do all of the following:

1. Explain the bug in simple terms.
2. Provide a corrected version of the code.
3. Suggest best practices or improvements.

Do NOT just repeat the code. Focus on explanation and fixes.

Code:
{code}

Explanation:
"""
    inputs = tokenizer(prompt, return_tensors="pt", padding=True).to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            max_new_tokens=250,
            do_sample=True,
            temperature=0.7,
            top_p=0.95,
            top_k=50,
            repetition_penalty=2.0
        )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
def review_code(code):
    bug_status = detect_bug(code)
    issue = "Bug"
    explanation = generate_explanation(code, issue)
    return bug_status, explanation
```

```python
sample_code = '''
def divide(a, b):
    return a / b

print(divide(10, 0))
'''

status, result = review_code(sample_code)
print(status)
print(result)

```

```python
import gradio as gr

def ui(code):
    status, explanation = review_code(code)
    return status, explanation

iface = gr.Interface(
    fn=ui,
    inputs=gr.Textbox(lines=12, placeholder="Paste your code here..."),
    outputs=[
        gr.Textbox(lines=2, label="Bug Status"),
        gr.Textbox(lines=15, label="Explanation & Fix")
    ],
    title="Smart Code Review Tool"
)

iface.launch()
```
